### (연료비, 전기차 증가율)로 상관관계를 분석
전기차 증가율 = (N년 전기차 대수 - (N-1)년 전기차 대수) / (N-1)년 전기차 대수 * 100

```python
all_car1 = (car_2020 - car_2019) / car_2019 * 100
all_car1
```

```python
all_car2 = (car_2021 - car_2020) / car_2020 * 100
all_car2
```

```python
all_car3 = (car_2022 - car_2021) / car_2021 * 100
all_car3
```

```python
all_data1 = all_car1, all_car2, all_car3
all_data = {'증가율': all_data1}

all_car = pd.DataFrame(all_data)
all_car
```
