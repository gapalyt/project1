```python
import requests
from bs4 import BeautifulSoup

url = "https://m.stock.naver.com/marketindex/energy/OIL_GSL"

res = requests.get(url)
page = res.content

soup = BeautifulSoup(page, 'html.parser')
soup
```

```python
url = "https://m.stock.naver.com/front-api/v1/chart/pricesByPeriod?category=energy&reutersCode=OIL_GSL&chartInfoType=marketindex&scriptChartType=areaYearFive"
res = requests.get(url)

datas = res.json()
datas
```

```python
datas = res.json()['result']['priceInfos']
datas
```

```python
import pandas as pd

OIL_GSL1 = pd.DataFrame(datas)
OIL_GSL1
```

```python
OIL_GSL_5 = OIL_GSL1.drop(['openPrice', 'closePrice'], axis = 1) # openPrice, closePrice 삭제
OIL_GSL_5
```

```python
OIL_GSL1_2019 = OIL_GSL_5[:]
OIL_GSL2_2019 = OIL_GSL1_2019.loc[(OIL_GSL1_2019['localDate'] <= "20191231") & (OIL_GSL1_2019['localDate'] >= "20190101")]
OIL_GSL2_2019['highPrice'].astype('float')
OIL_GSL2_2019['lowPrice'].astype('float')
OIL_GSL2_2019['price'] = (OIL_GSL2_2019['highPrice'] + OIL_GSL2_2019['lowPrice']) / 2
OIL_GSL2_2019['sum'] = OIL_GSL2_2019['price'].sum(axis = 0)
OIL_GSL2_2019['average'] = OIL_GSL2_2019['sum'].div(52)
```
