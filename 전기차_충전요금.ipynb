# 전기차 충전 요금 2019년

```python
import requests
from bs4 import BeautifulSoup

url = 'https://tech.hyundaimotorgroup.com/kr/article/all-about-evs-2-cost-of-a-full-charge/'

res = requests.get(url)
page = res.content

soup = BeautifulSoup(page, 'html.parser')
soup
```

```python
import re

korean = re.compile('[\u3131-\u3163\uac00-\ud7a3]+')

re_p_2019 = soup.select_one('p:nth-of-type(8)').text
p_2019 = re.sub(korean, '', re_p_2019) # 한글 제거
p_2019 = re.sub("[a-zA-Z]", "", p_2019) # 영어 제거
p_2019 = p_2019.strip() # 양쪽 공백 제거
p_2019 = p_2019[1:8] # 불필요한 문자 제거
p_2019 = p_2019.lstrip() # 왼쪽 공백 제거

p_2019
```

# 전기차 충전요금 2022년
